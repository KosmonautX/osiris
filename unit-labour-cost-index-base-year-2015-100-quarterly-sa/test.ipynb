```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("unit-labour-cost-index-base-year-2015-100-quarterly-sa.csv")
df.tail()
```

```python
# fig, ax = plt.subplots()
plt.figure(figsize=(25,14))
plt.plot(df.quarter, df.value)
plt.xlabel("Quarter", fontsize=20)
plt.ylabel("Unit Labour Cost of Overall Economy", fontsize=20)
plt.xticks(rotation=45)
plt.show()
```
